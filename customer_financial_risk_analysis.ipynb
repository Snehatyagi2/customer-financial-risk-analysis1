```python
import pandas as pd
```

```python
df = pd.read_csv("/content/marketing_campaign.csv", sep="\t")
```

```python
df = df.dropna()
```

```python
df.info()
```

```python
df["Total_Spending"] = (
    df["MntWines"] +
    df["MntMeatProducts"] +
    df["MntFishProducts"] +
    df["MntSweetProducts"] +
    df["MntGoldProds"]
)
```

```python
df["Spending_to_Income"] = df["Total_Spending"] / df["Income"]
```

```python
df["Spender_Type"] = df["Spending_to_Income"].apply(
    lambda x: "High Spender" if x > 0.6 else "Normal Spender"
)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x=df["Income"], y=df["Total_Spending"])
plt.title("Income vs Total Spending Behavior")
plt.show()
```

```python
df["Income_Group"] = pd.cut(
    df["Income"],
    bins=3,
    labels=["Low Income", "Middle Income", "High Income"]
)
```

```python
df["Spender_Class"] = df["Spending_to_Income"].apply(
    lambda x: "Overspender" if x > 0.6
    else "Balanced Spender" if x > 0.3
    else "Saver"
)
```

```python
df["Customer_Type"] = df["Income_Group"].astype(str) + " - " + df["Spender_Class"]
```

```python
df["Customer_Type"].value_counts()
```

```python
df.groupby("Customer_Type")["Total_Spending"].mean().sort_values(ascending=False)
```

```python
df.groupby("Income_Group")[["Total_Spending", "Spending_to_Income"]].mean()
```

```python
df["Risk_Score"] = (
    df["Spending_to_Income"] * 50 +
    df["Total_Spending"] / df["Total_Spending"].max() * 50
)
```

```python
def risk_level(x):
    if x > 70:
        return "High Risk"
    elif x > 40:
        return "Medium Risk"
    else:
        return "Low Risk"

df["Risk_Level"] = df["Risk_Score"].apply(risk_level)
```

```python
df.sort_values("Risk_Score", ascending=False).head(10)
```

```python
df[df["Total_Spending"] > df["Total_Spending"].quantile(0.95)]
```

```python
df[df["Spending_to_Income"] < 0.1]
```

```python
df["Alert"] = df["Risk_Level"].apply(lambda x: "Send Warning" if x == "High Risk" else "Normal")
```

```python
df["spending_norm"] = df["Total_Spending"] / df["Total_Spending"].max()
```

```python
df["Risk_Score"] = (
    df["Spending_to_Income"] * 0.7 +
    df["spending_norm"] * 0.3
) * 100
```

```python
def risk_label(x):
    if x >= 70:
        return "High Risk"
    elif x >= 40:
        return "Medium Risk"
    else:
        return "Low Risk"

df["Risk_Level"] = df["Risk_Score"].apply(risk_label)
```

```python
df["Risk_Level"].value_counts()
```

```python
df.groupby("Risk_Level")[["Income", "Total_Spending", "Spending_to_Income"]].mean()
```

```python
df.to_csv("customer_risk_dataset.csv", index=False)
```
